### 6. A normal hemoglobin A1C glucose level is below 5.7. Divide the data in 2 groups (patients that have normal A1C level and patients that have elevated A1C level), then sketch the data using a plot of your choice dependent on the length of the disease. Distinguishing both groups.

```python
patient_with_normal_A1C = df[(df['First-trimester (hemoglobin A1C) sugar'] <=5.7)]
```

```python
patient_with_elevated_A1C = df[(df['First-trimester (hemoglobin A1C) sugar'] >5.7)]
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
patient_with_normal_A1C.plot.scatter(x='How Long is a Diabetic', y='First-trimester (hemoglobin A1C) sugar', c='seagreen',label='patient_with_normal_A1C', ax=ax)
patient_with_elevated_A1C.plot.scatter(x='How Long is a Diabetic', y='First-trimester (hemoglobin A1C) sugar',c='crimson',label='patient_with_elevated_A1C', ax=ax)
plt.title('Patient with normal A1C vs Patient with elevated A1C')
plt.xlabel('How Long is a Diabetic')
plt.ylabel('First-trimester (hemoglobin A1C) sugar')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(15,9))
sns.set_style("whitegrid")
sns.boxplot(x='How Long is a Diabetic', y='First-trimester (hemoglobin A1C) sugar', data=patient_with_normal_A1C)
sns.boxplot(x='How Long is a Diabetic', y='First-trimester (hemoglobin A1C) sugar', data=patient_with_elevated_A1C)
plt.title('Patient_with_normal_A1C vs Patient_with_elevated_A1C')
plt.xlabel('How Long is a Diabetic')
plt.ylabel('First-trimester (hemoglobin A1C) sugar')
plt.xticks(rotation=45)
plt.show()
```

### 7. How do factors influence the outcome whether a patient will have (or not) diabetic retinopathy? Analyse whether there is a correlation or not between diabetic retinopathy and other data in the set. Do this by separating the patients into 2 groups and analyzing the characteristics of the groups with respect to various information. Plot various data points if possible.

```python
df['Does he have diabetic retinopathy?'].value_counts()
```

```python
plt.figure(figsize=(20,10))
y = sns.countplot(x="How Long is a Diabetic", hue="Does he have diabetic retinopathy?", data=df)
for container in y.containers:
    y.bar_label(container)
plt.title('Does he have diabetic retinopathy or not')
plt.show()
```

```python
# Spored prikazaniot plot,mozeme da zaklucime deka pacientot sto poveke godini ima dijabet, poveke e verojatno istiot
# da dobie i dijabeticna retinopatija, no sepak istoto zavisi i od mnogu drugi faktori, kako na primer kolku 
# samiot organizam e osteten od visokata glikoza vo krvta,kolku istiot e otporen i na sekoj pacient vlijae razlicno.
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x="Does he have diabetic retinopathy?", y="Glucose (current) in the blood", data=df)
plt.title('Diabetic retinopathy vs Glucose in the blood ')
plt.show()
```
